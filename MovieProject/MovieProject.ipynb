```python
# Import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjust the configuration of the plots we will create

# Read in the data
df = pd.read_csv(r"C:\Users\WAlvi\Documents\projects\Alex the A\Proj 2 python\movies.csv\movies.csv")
```

```python
# Let's look at the data

df.head()
```

```python
# Check for missing data or NULL
for col in df.columns:
    print(df[col].isnull().value_counts(), "\n")
```

```python
# Drop NULLs
df = df.dropna(how='any',axis=0) 
```

```python
# Change data type of budget & votes & gross to remove .0
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
# Check that it worked
df.head()
```

```python
# Remove words from released and turn into datatime
df['release_date'] = df.apply(lambda x: x['released'][0:x['released'].find(' (')],axis=1)

df['release_date'] = pd.to_datetime(df['release_date'], infer_datetime_format=True)

### I also changed the data in released to the correct format
##df['released'] = df.apply(lambda x: x['released'][0:x['released'].find(' (')],axis=1)

##df['released'] = pd.to_datetime(df['released'], infer_datetime_format=True)
```
